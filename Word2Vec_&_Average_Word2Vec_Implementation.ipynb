Data Cleaning & Preprocessing

```python
import re       # Regular Expression
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
from nltk.corpus import stopwords

corpus=[]
for i in range(0,len(messages)):
  review=re.sub("[^a-zA-Z0-9]"," ",messages["message"][i])
  review=review.lower()
  review=review.split()

  review=[ps.stem(word) for word in review if not word in stopwords.words("english")]
  review=" ".join(review)
  corpus.append(review)
```

```python
corpus
```

Initializing the Bag of Words Model

```python
from sklearn.feature_extraction .text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(2,2),binary=True)
X=cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
X.shape
```

```python
y=pd.get_dummies(messages["label"])
y=y.iloc[:,1].values
```

```python
y
```

Splitting the data into Training & Testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
X_train,y_train
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,y_pred).mean()*100
print(score)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(classification_report(y_pred,y_test))
```

TF - IDF Vector Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2700,ngram_range=(1,2),binary=True)
X=tv.fit_transform(corpus).toarray()
```

Splitting the Data into Training & Testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
```

```python
score
```

# **Word 2 Vec Implementation**

```python
!pip install gensim
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords

corpus=[]
for i in range(0,len(messages)):
  review=re.sub("[^a-zA-Z0-9]"," ",messages["message"][i])
  review=review.lower()
  review=review.split()

  review=[lemmatizer.lemmatize(word) for word in review if not word in stopwords.words("english")]
  review=" ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
corpus[8]
```

```python
import nltk
nltk.download('punkt')
```

```python
words=[]
for sent in corpus:
  sent_token=sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
```

```python
model=gensim.models.Word2Vec(words,window=5,min_count=2)
```

```python
model.wv.index_to_key
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.wv.similar_by_word("bold")
```

```python
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```
